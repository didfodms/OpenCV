```python
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = cv2.imread('images.bmp') #BGR 
plt.imshow(img1) #RGB 
#    
```

```python
plt.axis('off') #()   
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) #cvt = convert, BGRRGB   
plt.imshow(img2)
```

```python
imgGray = cv2.imread('images.bmp', cv2.IMREAD_GRAYSCALE) # grayscale 
```

```python
#grayscale imshow  
cv2.namedWindow('image')
cv2.imshow('image', imgGray)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(imgGray) #plt    .
```
