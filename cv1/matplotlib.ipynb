```python
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = cv2.imread('images.bmp') #BGR 순서
plt.imshow(img1) #RGB 순서
#그래서 색깔이 이상하게 나옴 
```

```python
plt.axis('off') #축(눈금) 사라지게 하기 
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) #cvt = convert, BGR에서RGB로 변환한다는 뜻 
plt.imshow(img2)
```

```python
imgGray = cv2.imread('images.bmp', cv2.IMREAD_GRAYSCALE) #이미지를 grayscale로 불러오겠다
```

```python
#grayscale을 imshow 창으로 띄우기
cv2.namedWindow('image')
cv2.imshow('image', imgGray)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(imgGray) #plt로 띄우면 이상한 색이 나온다.
```

```python
import numpy as np
z = np.array([[10, 0, 0], [0, 10, 0], [0, 0, 10]])
```

```python
plt.imshow(z) #plt자체의 기본색이 노란색 보라색이라서 이상한 색이 나옴.
```

```python
plt.imshow(z, cmap = 'gray') #이렇게 해주면 검흰으로 나옴 
```

```python
plt.imshow(imgGray, cmap = 'gray') #plt로 grayscale 출력하기!!
```

```python
plt.subplot(121) #1은 세로 비율, 2는 가로 비율, 위치 1로 설정해서 코딩
plt.imshow(img2)
plt.subplot(122) #1은 세로 비율, 2는 가로 비율, 위치 2로 설정해서 코딩
plt.imshow(imgGray, cmap='gray')
plt.show()
```
